# Look into the data

### 通过简单的数据可视化化手段，挖掘数据的信息

* 查看特征自身的情况

* 查看特征之间的关系


```python
import pandas as pd
# 绘图
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python
df = pd.read_excel('C:/Working Tem/classes/realestate_sample.xlsx')
df.head()
```

```python
# 制作类别标签
price_median = df['average_price'].median()
df['is_high'] = df['average_price'].map(lambda x: 'True' if x>= price_median else 'False')
```

### 综合查看

```python
# 查看数据列名
df.columns
```

```python
# 查看数据缺失情况
df.isnull().sum()
```

```python
# 查看数据的数据类型
df.dtypes
```

### 连续变量查看

#### 查看数据频数分布情况

```python
# 查看数据分布
df['average_price'].hist()
```

#### 查看数据异常值分布情况

```python
# 查看数据异常值情况
df[['average_price']].boxplot()
```

```python
# 箱线图支持多个变量同事查询
df[['daypop','nightpop']].boxplot()
```

```python
# 多变量查看箱线图，要注意特征的量纲
df[['daypop','nightpop','sub_kde']].boxplot()
```

#### 查看变量间的关系

一般散点图

```python
df.plot.scatter(x='area',y='average_price')
```

高级一点的分布+散点图+趋势线

```python
ax = sns.jointplot(x='area',y='average_price', data=df, kind='reg')
```

一口气查看多个变量间的散点图与数据分布图

```python
sns.pairplot(df[[ 'average_price', 'area', 'daypop', 'nightpop', 'sub_kde', 'bus_kde']],kind="reg")
```

分类查看变量间的散点图

```python
g = sns.pairplot(df[[ 'average_price', 'area', 'daypop', 'nightpop', 'sub_kde', 'bus_kde','is_high']], hue='is_high', markers=["o", "s"])
```

查看变量的相关性矩阵

```python
# 查看特征相关性矩阵
df_draw = df[[ 'average_price', 'area', 'daypop', 'nightpop', 'sub_kde', 'bus_kde']]
plt.figure(figsize=(24,12))
sns.heatmap(df_draw.corr(),linewidths=0.1,vmax=1.0, 
            square=True, linecolor='white', annot=True)
plt.show()
```

### 离散变量查看

#### 查看频数

频数分布

```python
df.is_high.value_counts()
```

频数柱状图

```python
colors = ["#191970", "#DC143C"]
sns.countplot('is_high', data=df, palette=colors)
```

#### 查看类别与连续变量之间的关系

分类别查看数据散点分布

```python
ax = sns.swarmplot(x='is_high', y='average_price', data=df)
```

分类别查看数据分布

```python
df = df.fillna(0)
sns.distplot(df[df['is_high'] == 'False']["average_price"], color='#4169E1') 
sns.distplot(df[df['is_high'] == 'True']["average_price"], color='#DC143C') 
```

一口气查看多个特征的分类别分布情况

```python
columns = ['average_price', 'area', 'daypop', 'nightpop', 'sub_kde', 'bus_kde']
grid = gridspec.GridSpec(3, 2)
plt.figure(figsize=(15,6*4))
for n, col in enumerate(df[columns]):
    ax = plt.subplot(grid[n])
    sns.distplot(df[col][df['is_high'] == 'False'], bins = 50, color='#DC143C') # 红色
    sns.distplot(df[col][df['is_high'] == 'True'], bins = 50, color='#4169E1') # 蓝色
    ax.set_ylabel('Density')
    ax.set_title(str(col))
    ax.set_xlabel('')
plt.show()
```
